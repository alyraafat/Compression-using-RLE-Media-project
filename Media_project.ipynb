##Given Helper Functions##

```python
def get_size(image):
  return image.size
```

```python
def quantization(image, n):
  return image.quantize(n)
```

**Modified Cell**

```python
def get_pixel_value(img,col,row):
  return img.getpixel((col,row))
```

```python
def get_unique_values(image):
  return np.unique(image), len(np.unique(image))
```

```python
def extract_row_color(arr,row,color):
  filtered = filter(lambda item: ((item[0] == row) and (item[-1] == color)), arr)
  return list(filtered)
```

```python
def merge_row_color(filtered, row, color):
  remove_row_color = [item[1:-1] for item in filtered]
  merge_runs = (row,)
  for item in remove_row_color:
    merge_runs = merge_runs + item
  otp = merge_runs + (color,)
  return otp
```

##Task 1##

```python
def show_image_information(image):
  original_colors = []
  quantized_colors=[]
  # ......start code......
  original_colors = get_unique_values(image)[0]
  quantized_image = quantization(image,9)
  quantized_colors = get_unique_values(quantized_image)[0]
  # ......end code......
  return list(original_colors),list(quantized_colors)
```

##Task 2##
**Task 2: Compute the runs per each row for the image**

```python
def compute_runs(image, unique_values):
  runs_code = []
  # ......start code......
  i=0
  size = get_size(image)
  while i<size[1]: # loop over row (height)
    #for val in unique_values:
    temp = [-1,-1]
    j=0
    while j<size[0]-1: # loop over col (width)
      pixel = get_pixel_value(image,j,i)
      next_pixel = get_pixel_value(image,j+1,i)
      if temp[0]==-1:
        temp[0] = j
        temp[1] = j
      if pixel==next_pixel and temp[0]!=-1:
        temp[1] = j
      if pixel!=next_pixel and temp[0]!=-1:
        temp[1] = j
        runs_code+=[tuple([i]+temp+[pixel])]
        temp = [-1,-1]
      j+=1
    pixel = get_pixel_value(image,j,i)
    t = runs_code[len(runs_code)-1] 
    if temp[0]!=-1:
      temp[1]=j
      runs_code+=[tuple([i]+temp+[pixel])]
    else:
      runs_code+=[tuple([i]+[j,j]+[pixel])]
    i+=1
  # ......end code......
  return runs_code
```

##Task 3##
**Compute the Run-Length Encoding of an Image**

```python
def compute_RLE(image, unique_values):
  result = []
  # ......start code......
  runs_code = compute_runs(image, unique_values)
  i=0
  size = get_size(image)
  while i<size[1]:
    for color in unique_values:
      filtered = extract_row_color(runs_code,i,color)
      if len(filtered)>0:
        otp = merge_row_color(filtered, i, color)
        result+=[otp]
    i+=1
  # ......end code......
  return result
```
